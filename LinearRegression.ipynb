```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {
    'Area': [1000,1200,1500,1800,2000,2500],
    'Price': [50,60,75,90,100,125]
}

df = pd.DataFrame(data)

X = df[['Area']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)

new_house_area = [[2200]]

predicted_price = model.predict(new_house_area)

print("House Area:", new_house_area[0][0], "sq.ft")
print("Predicted House Price:",round(predicted_price[0],2),"Lakhs")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


data = {
    'Experiance': [3,5,8,10,12,15],
    'Salary': [60000,70000,80000,90000,120000,150000]
}

df = pd.DataFrame(data)

x = df[['Experiance']]
y = df['Salary']
plt.xlabel("Experiance")
plt.ylabel("Salary")
model = LinearRegression()
model.fit(x, y)

years = float(input("Enter the years of experiance: "))

new_experiance = [[years]]
predicted_salary = model.predict(new_experiance)

print("Experiance:", new_experiance[0][0], "Years")
print("Predicted Salary", round(predicted_salary[0],2), "Rupees")
plt.plot(x,y)
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
data = {
    "Area": [1000,1200,1500,1800,2000,2500],
    "Bedrooms": [2,2,3,3,4,4],
    "Age": [10,8,5,7,3,2],
    "Price":[50,60,75,90,100,125]
}

df = pd.DataFrame(data)
x = df[["Area","Bedrooms","Age"]]
y = df["Price"]
plt.plot(x,y)
plt.show()
model = LinearRegression()
model.fit(x,y)

new_house = [[2200,4,4]]
predicted_price = model.predict(new_house)

print("Predicted House Price:", round(predicted_price[0],2),"Lakhs")
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

data = {
    "Hours": [1,2,3,4,5,6],
    "Pass": [0,0,0,1,1,1]
}
df = pd.DataFrame(data)

x = df[["Hours"]]
y = df["Pass"]

model = LogisticRegression()
model.fit(x,y)

prediction = model.predict([[2.5]])

if prediction[0]==1:
  print("Student Pass")
else:
  print("Student Fail")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    "temp": [30,35,36,41,42,46,48],
    "sales": [1,1.2,1.5,2.2,2.3,2.7,3.5]
}

df = pd.DataFrame(data)
x = df[["temp"]]
y = df["sales"]

model = LinearRegression()
model.fit(x,y)

new_temp = [[50]]

predicted_price = model.predict(new_temp)

print("New Temperature:", new_temp[0][0], "Degree C")
print("Predicted Ice Cream Sales:",round(predicted_price[0],2),"Lakhs")
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

data = {
    "Marks": [25, 32, 38, 45, 65, 85],
    "Pass": [0, 0, 0, 1, 1, 1]
}

df = pd.DataFrame(data)

# plt.scatter(df["Marks"],df["Pass"])

x = df[["Marks"]]
y = df["Pass"]



model = LogisticRegression()
model.fit(x, y)
prediction = model.predict([[42]])

plt.plot(x,y)
plt.xlabel("Marks")
plt.ylabel("Pass")
plt.show()
if prediction[0] == 1:
    print("Student Pass")
else:
    print("Student Fail")


```
